# Step 1
### Use Pandas to load Abalone Data Set as the  dataframe from UCI Machine Learning Repository.  
**[Pandas](https://pandas.pydata.org/)**  
**[Abalone Data Set](https://archive.ics.uci.edu/ml/datasets/Abalone)**

```python
import pandas as pd
import random

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data', 
                 header=None,
                 names=['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings'])
```

# Step 6
### Plot the distribution of three types of Abalone using different colors where x-axis is the diameter and y-axis the whole weight

```python
# df.head(20)

df_M = df[(df.Sex == "M")]
df_M_sorted = df_M.sort_values("Whole weight",ascending= True)
# print(df_M_sorted.loc[:]["Whole weight"])
# print(df_M_sorted.loc[:]["Diameter"])

df_F = df[(df.Sex == "F")]
df_F_sorted = df_F.sort_values("Whole weight",ascending= True)

df_I = df[(df.Sex == "I")]
df_I_sorted = df_I.sort_values("Whole weight",ascending= True)

plt.xlabel('Whole weight')
plt.ylabel('Diameter')
plt.scatter(df_M_sorted.loc[:]["Whole weight"], df_M_sorted.loc[:]["Diameter"],c = 'b',marker = 'o')
plt.scatter(df_F_sorted.loc[:]["Whole weight"], df_F_sorted.loc[:]["Diameter"],c = 'r',marker = 'o')
plt.scatter(df_I_sorted.loc[:]["Whole weight"], df_I_sorted.loc[:]["Diameter"],c = 'g',marker = 'o')


fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax1.scatter(df_M_sorted.loc[:]["Whole weight"], df_M_sorted.loc[:]["Diameter"],c = 'b',marker = 'o')
ax2.scatter(df_F_sorted.loc[:]["Whole weight"], df_F_sorted.loc[:]["Diameter"],c = 'r',marker = 'o')
ax3.scatter(df_I_sorted.loc[:]["Whole weight"], df_I_sorted.loc[:]["Diameter"],c = 'g',marker = 'o')

plt.show()
```

# Step 7
### Show the proportion of three types of Abalone using pie chart.

```python
labels = 'Men', 'Female', 'Infant'

count_M = df[(df.Sex == "M")].shape[0]
count_F = df[(df.Sex == "F")].shape[0]
count_I = df[(df.Sex == "I")].shape[0]
size = [count_M, count_F, count_I]
print("Men : " + str(count_M))
print("Female : " + str(count_F))
print("Infant : " + str(count_I))

plt.pie(size, labels = labels, autopct = '%1.1f%%')
plt.axis('equal')
plt.show();
```
